```python
import matplotlib.pyplot as plt
import math as m

def generator_data(a, b, size):
    n = (b - a) / (size - 1)
    result = []
    s = a
    
    while s < b:
        result.append(s)
        s = s + n
    if len(result) < size:
        result.append(b)
    
    return result

X = generator_data(4, 6, 100)
print(X)
```

```python
def pmf_normal(x, mu, sigma):
    return (1 / m.sqrt(2 * m.pi * sigma**2)) * m.exp(-1 * ((x - mu)**2 / (2 * sigma**2)))

def cdf_normal(x, mu, sigma):
    return 0.5 * (1 + m.erf((x - mu) / (sigma * m.sqrt(2))))
```

```python
def plot_pmf_normal(mu, sigma):
    X = generator_data(mu - 4*sigma, mu + 4*sigma, 1000)
    P_normal = [pmf_normal(x, mu, sigma) for x in X]
    plt.plot(X, P_normal, '-')
    plt.title('PMF of Normal(%.2f, %.2f)' %(mu, sigma))
    plt.xlabel('X')
    plt.ylabel('P')
    plt.show()
    
plot_pmf_normal(0, 1.5)
```

```python
def plot_cdf_normal(mu, sigma):
    X = generator_data(mu - 4*sigma, mu + 4*sigma, 1000)
    P_normal = [cdf_normal(x, mu, sigma) for x in X]
    plt.plot(X, P_normal, '-')
    plt.title('CDF of Normal(%.2f, %.2f)' %(mu, sigma))
    plt.xlabel('X')
    plt.ylabel('P')
    plt.show()
    
plot_cdf_normal(0, 1.5)
```

```python
mu, sigma = 10, 1
result = cdf_normal(12, mu, sigma) - cdf_normal(9, mu, sigma)
print('The product will be generated in the period from 9 minutes to 12 minutes: %.4f' %(result))
```
