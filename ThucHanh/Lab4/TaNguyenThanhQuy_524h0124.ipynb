```python
import random
import math
import matplotlib.pyplot as plt

x = [random.randint(1,6) for i in range(8000)]
X = set(x)
print (x)
print (X)

P = [x.count(i)/len(x) for i in X]
print (P)

FX = sum(P[:3])
print (FX)

EX = 0
for x in X:
    EX = EX + (x * P[x-1]);
print (EX)

VarX = 0
for x in X:
    VarX = VarX + (x-EX) * (x-EX) * P[x-1]
print (VarX)

SD = math.sqrt(VarX)
print (SD)
```

```python
# Ex1
n = 10000
rolls = [random.randint(1, 6) for _ in range(n)]

counts = [rolls.count(i) for i in range(1, 7)]

pmf = [c / n for c in counts]

for i, p in zip(range(1, 7), pmf):
    print(f"P(X={i}) = {p:.4f}")

plt.bar(range(1, 7), pmf, color='lightblue', edgecolor='black')
plt.xlabel("Face of Die (X)")
plt.ylabel("PMF")
plt.title("Probability Mass Function")
plt.show()
```

```python
# Ex2
from scipy.stats import norm
import numpy as np

mu = 0
sigma = 1
dist = norm(loc=mu, scale=sigma)

x = np.linspace(-4, 4, 200)

pdf = dist.pdf(x)

plt.plot(x, pdf, color='blue', linewidth=2, label='PDF of N(0,1)')
plt.title('Normal Distribution (=0, =1)')
plt.xlabel('x')
plt.ylabel('Probability Density f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Ex3
a, b = 0, 1

x = np.linspace(0, 1, 100)

F = (x - a) / (b - a)

plt.plot(x, F, color='red', linewidth=2, label='CDF of Uniform(0,1)')
plt.title('Cumulative Distribution Function for Uniform(0,1)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()
plt.show()
```
