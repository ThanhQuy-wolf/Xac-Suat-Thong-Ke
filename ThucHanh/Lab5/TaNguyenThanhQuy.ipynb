```python
import random

def randomTwoDiceSame():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a == b:
            count += 1;
            
    return count / n;

print("Both dice are the same:", randomTwoDiceSame())
```

```python
def randomTwoDiceDiff():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a != b:
            count += 1;
            
    return count / n;

print("Both dice are the different:", randomTwoDiceDiff())
```

```python
def randomTwoDiceEven():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a % 2 == 0 and b % 2 == 0:
            count += 1;
            
    return count / n;

print("Both dice are even:", randomTwoDiceEven())
```

```python
def randomTwoDiceOdd():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a % 2 != 0 and b % 2 != 0:
            count += 1;
            
    return count / n;

print("Both dice are odd:", randomTwoDiceOdd())
```

```python
def randomEvenOdd():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if (a % 2 == 0 and b % 2 != 0) or (a % 2 != 0 and b % 2 == 0):
            count += 1;
            
    return count / n;

print("One dice is even and the other is odd:", randomEvenOdd())
```

```python
def randomTwoDiceAre6():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a == 6 and b == 6:
            count += 1;
            
    return count / n;

print("Both dice are 6:", randomTwoDiceAre6())
```

```python
def sumGreaterThan10():
    n = 10000;
    count = 0;
    for _ in range(n):
        a, b = random.randint(1, 6), random.randint(1, 6);
        if a + b > 10:
            count += 1;
            
    return count / n;

print("Summation of both dice are greater than 10:", sumGreaterThan10())
```

```python
def allSameColor():
    n = 10000
    balls = ['blue'] * 2 + ['red'] * 3 + ['white'] * 5
    count = 0
    
    for _ in range(n):
        pick = random.sample(balls, 3)
        if len(set(pick)) == 1:
            count += 1
    
    return count / n   

print("All 3 balls are same color:", allSameColor())
```

```python
def twoRedOneWhite():
    n = 10000
    balls = ['blue'] * 2 + ['red'] * 3 + ['white'] * 5
    count = 0
    
    for _ in range(n):
        pick = random.sample(balls, 3)
        if pick.count('red') == 2 and pick.count('white') == 1:
            count += 1;
    
    return count / n   

print("There are 2 red balls and 1 white ball:", twoRedOneWhite())
```

```python
import itertools as itp

def listThreeWhiteBall():
    balls = ['B1', 'B2', 'R1', 'R2', 'R3', 'W1', 'W2', 'W3', 'W4', 'W5']
    
    white_balls = [b for b in balls if b.startswith('W')]
    
    combinations_white = list(itp.combinations(white_balls, 3))
    
    print("List all the cases that all 3 balls are white")
    for i, combo in enumerate(combinations_white, start=1):
        print(f"{i}. {combo}")
        
listThreeWhiteBall()
```

```python
def buildDeck():
    suits = ['C', 'D', 'H', 'S']
    ranks = ['A'] + [str(i) for i in range(2, 11)] + ['J', 'Q', 'K']
    deck = []
    for s in suits:
        color = 'red' if s in ('D', 'H') else 'black'
        for r in ranks:
            deck.append({'rank': r, 'suit': s, 'color': color})
    
    return deck

DECK = buildDeck()
```

```python
def allSameSuit():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if len({c['suit'] for c in pick}) == 1:
            count += 1
            
    return count / n

print("All same suit:", allSameSuit())
```

```python
def allDiffSuit():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if len({c['suit'] for c in pick}) == 4:
            count += 1
            
    return count / n

print("All different suit:", allDiffSuit())
```

```python
def allSameColorDeck():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if len({c['color'] for c in pick}) == 1:
            count += 1
            
    return count / n

print("All same color in deck:", allSameColorDeck())
```

```python
def allSameRank():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if len({c['rank'] for c in pick}) == 1:
            count += 1
            
    return count / n

print("All same rank:", allSameRank())
```

```python
def allIsDigit():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if all(c['rank'].isdigit() for c in pick):
            count += 1
            
    return count / n

print("All is digit:", allIsDigit())
```

```python
def allIsFace():
    n = 10000
    count = 0
    for _ in range(n):
        pick = random.sample(DECK, 4)
        if all(c['rank'] in {'J', 'Q', 'K', 'A'} for c in pick):
            count += 1
            
    return count / n

print("All is face:", allIsFace())
```
