```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

n = 5
p = 0.1

X = np.arange(0, n + 1)

P = [comb(n,k) * (p**k) * ((1 - p) ** (n - k)) for k in X]

for k, prob in zip(X, P):
    print(f"P(X={k}) = {prob:.3f}")


plt.plot(X, P, '-o')
plt.title('Binomial Distribution: n=5, p=0.1')
plt.xlabel('S my hng (X)')
plt.ylabel('Xc sut P(X)')
plt.grid(True)
plt.show()
```

```python
import math

lamda = 3
X = np.arange(1, 6)

# P(X=k) = e^(-) * ^k / k!
P = [math.exp(-lamda) * lamda**k / math.factorial(k) for k in X]

for k, p in zip(X, P):
    print(f"P(X={k}) = {p:.3f}")

plt.plot(X, P, '-o')
plt.title('Poisson Distribution (=3)')
plt.xlabel('S cuc gi (X)')
plt.ylabel('Xc sut P(X)')
plt.grid(True)
plt.show()
```

```python
p = 0.4

X = np.arange(1, 11)

# P(X = k) = (1 - p)^k-1 * p
P = (1 - p) ** (X - 1) * p

for k, prob in zip(X, P):
    print(f"P(X={k}) = {prob:.3f}")

plt.plot(X, P, 'o-', color='royalblue')
plt.title('PMF of Geometric(p=0.4)')
plt.xlabel('Shot number (X)')
plt.ylabel('Probability P(X)')
plt.grid(True)
plt.show()
```
