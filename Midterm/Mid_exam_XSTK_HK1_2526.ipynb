# Question 1: (2 points)
A closed urn has 23 balls: 8 white, 6 blue and 9 red.

````python
def cross(A, n):
    return {a+str(b+1) for a in A for b in range(n)}

urn = cross('W', 8) | cross('B', 6) | cross('R', 9)
````

Randomly select ***x*** balls
(know that each ball has the same probability of selection).

For example: Student ID: 0412347 => 47

````python
def get_x(n):
    return int(( n / 99 ) * 4 + 6)

x = get_x(24)
````

Find a list of all possible ***x*** balls and assign to variable $U$.

````python
import itertools as itt

U = set(itt.combinations(urn, x))

print(f"Number of ways to select {x} balls: {len(U)}")
print(f"First 5 combinations: {list(U)[:5]}")
````

Let *X* represent the random variable corresponding to the number of red balls in ***x*** balls.

Let *Y* represent the random variable corresponding to the number of white balls in ***x*** balls.

Compute the probability:
$P(X=3|Y=1)$ (present the steps along with illustrative code).

````python
# Công thức: P(X=3|Y=1) = count / total
# count = P(X=3 ∩ Y=1)
# total = P(Y=1)

def countXY(U):
    count = 0
    total = 0
    
    for combo in U:
        reds = sum('R' in c for c in combo)
        whites = sum('W' in c for c in combo)
        if whites == 1 and reds == 3:
            count+=1
        if whites == 1:
            total+=1
    return count / total if total else 0

P = countXY(U)
print(f"P(X=3 | Y=1) = {P:.5f}")
````

**2.1 (0.5 point):** Consider a random variable that follows normal distribution with mean 6 and variance 12.

Calculate: $P(4 \leq X \leq 8)$

**Write your answer here...**


````python
from scipy.stats import norm
import math

# Function prototype:
# norm.cdf(x, loc=mean, scale=std)
# → Input: x (value), loc (mean μ), scale (standard deviation σ)
# → Output: P(X ≤ x)

mu = 6
sigma = math.sqrt(12)

P = norm.cdf(8, loc=mu, scale=sigma) - norm.cdf(4, loc=mu, scale=sigma)
print(f"P(4 ≤ X ≤ 8) = {P:.4f}")
````
